```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv((r'../data/Salary_Data.csv'))
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Age'] = df['Age'].astype('Int64')
df['Salary'] = df['Salary'].astype('Int64')
```

```python
df.sample(5)
```

An√°lises Univariadas

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(14,5))

sns.histplot(data=df, x='Age', bins=10,ax=ax1,legend=False,kde=True)
ax1.set_title('Histograma Idades')
ax1.set_xlabel('Idade')
ax1.set_ylabel('Quantidade')

sns.histplot(data=df, x='Years of Experience', bins=10,ax=ax2,legend=False,kde=True)
ax2.set_title('Histograma Anos Experi√™ncia')
ax2.set_xlabel('Tempo')
ax2.set_ylabel('Quantidade')

sns.histplot(data=df, x='Salary', bins=10,ax=ax3,legend=False,kde=True)
ax3.set_title('Histograma Sal√°rios')
ax3.set_xlabel('Sal√°rio')
ax3.set_ylabel('Quantidade')

plt.tight_layout()
plt.show()
```

```python
df.sample(5)
```

```python
# 10 Maiores m√©dias de sal√°rios

top_media_salarios = df.groupby('Job Title')['Salary'].mean().sort_values(ascending=False).head(10).reset_index()


plt.figure(figsize=(6,4))
sns.barplot(data=top_media_salarios,x='Salary',y='Job Title',hue='Job Title',palette='mako')
plt.title('10 maiores m√©dias de sal√°rios')
plt.show()
```

```python
# Scatterplot relacionando os anos de experiencia com o sal√°rio

plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x='Salary',y='Age',alpha=0.3)
plt.title('Rela√ß√£o Exp x Sal√°rio')
plt.show()
```

```python
sns.scatterplot(data=df, x='Salary', y='Age', alpha=0.3)
```

```python
plt.figure(figsize=(8, 6))
# Usa histplot com x e y para criar um mapa de calor 2D
sns.histplot(data=df, x='Salary', y='Age', bins=50, cmap="viridis", cbar=True)

plt.title('Rela√ß√£o Idade x Sal√°rio (Densidade 2D)')
plt.xlabel('Sal√°rio')
plt.ylabel('Idade')
plt.show()
```
