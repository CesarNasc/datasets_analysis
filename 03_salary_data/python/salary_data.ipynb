```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv((r'../data/Salary_Data.csv'))
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Age'] = df['Age'].astype('Int64')
df['Salary'] = df['Salary'].astype('Int64')
```

```python
df.sample(5)
```

Estat√≠sticas Univariadas

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(14,5))

sns.histplot(data=df, x='Age', bins=10,ax=ax1,legend=False,kde=True)
ax1.set_title('Histograma Idades')
ax1.set_xlabel('Idade')
ax1.set_ylabel('Quantidade')

sns.histplot(data=df, x='Years of Experience', bins=10,ax=ax2,legend=False,kde=True)
ax2.set_title('Histograma Anos Experi√™ncia')
ax2.set_xlabel('Tempo')
ax2.set_ylabel('Quantidade')

sns.histplot(data=df, x='Salary', bins=10,ax=ax3,legend=False,kde=True)
ax3.set_title('Histograma Sal√°rios')
ax3.set_xlabel('Sal√°rio')
ax3.set_ylabel('Quantidade')

plt.tight_layout()
plt.show()
```
