```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/Coffe_sales.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S.%f', errors='coerce') 
```

```python
df.info()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Monthsort',bins=12,kde=True,edgecolor='black')
plt.title('Histograma Meses')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# Dias da semana vão de 1 a 7 (1 = segunda / 7 = domingo)

plt.figure(figsize=(6,4))
sns.histplot(data=df, x='Weekdaysort',kde=True,edgecolor='black')
plt.title('Histograma Dias da Semana')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='hour_of_day',kde=True,edgecolor='black')
plt.title('Histograma Horários de Vendas')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
